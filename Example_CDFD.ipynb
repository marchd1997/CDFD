# Example usage decomposition_helpers

```python
import numpy as np
import pandas as pd
import networkx as nx
import time
import os

import matplotlib.pyplot as plt

from decomposition_helpers import get_circularity, CDFD
from other_helpers import  uniform_multigraph, in_cycle_ratio, trophic_incoherence
```

## Example network

```python
G = nx.DiGraph() 
G.add_weighted_edges_from([(1, 2, 1), (2, 3, 1), (1,3,1),(3,1,1)])
pos={1: np.array([0,1]), 2: np.array([-0.3,0]), 3: np.array([0,-1])}# pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.show()
```

### BFF

```python
C, D = CDFD(G,solution_method = "BFF")
print("Circularity ratio is", get_circularity(C,D))
print("The directional part is:")
nx.draw_networkx(D,pos)
labels = nx.get_edge_attributes(D,'weight')
nx.draw_networkx_edge_labels(D,pos,edge_labels=labels)
plt.show()
```

### Maximal

```python
C, D = CDFD(G,solution_method = "Maximal")
print("Circularity ratio is", get_circularity(C,D))
print("The directional part is:")
nx.draw_networkx(D,pos)
labels = nx.get_edge_attributes(D,'weight')
nx.draw_networkx_edge_labels(D,pos,edge_labels=labels)
plt.show()
```
