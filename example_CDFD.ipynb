# Example usage decomposition_helpers

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from CDFD import get_circularity, get_directedness, CDFD
```

## Example network

```python
G = nx.DiGraph() 
G.add_weighted_edges_from([(1, 2, 1), (2, 4, 1), (1,3,1), (3,4,1), (4,1,1)])
pos={1: np.array([0,1]), 2: np.array([-1,0]), 3: np.array([1,0]), 4: np.array([0,-1])}# pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.show()
```

### BFF

```python
C, D = CDFD(G,solution_method = "BFF")

print("Circularity ratio is", get_circularity(C,D))
print("The circular part is:")
nx.draw_networkx(C,pos)
labels = nx.get_edge_attributes(C,'weight')
nx.draw_networkx_edge_labels(C,pos,edge_labels=labels)
plt.show()

print("Directedness ratio is", get_directedness(C,D))
print("The directional part is:")
nx.draw_networkx(D,pos)
labels = nx.get_edge_attributes(D,'weight')
nx.draw_networkx_edge_labels(D,pos,edge_labels=labels)
plt.show()
```

Notice that *circularity + directedness = 1*.

The weight between *3* and *4* suffers from a small numerical error. 

### Maximal

```python
C, D = CDFD(G,solution_method = "Maximal")

print("Circularity ratio is", get_circularity(C,D))
print("The circular part is:")
nx.draw_networkx(C,pos)
labels = nx.get_edge_attributes(C,'weight')
nx.draw_networkx_edge_labels(C,pos,edge_labels=labels)
plt.show()

print("Directedness ratio is", get_directedness(C,D))
print("The directional part is:")
nx.draw_networkx(D,pos)
labels = nx.get_edge_attributes(D,'weight')
nx.draw_networkx_edge_labels(D,pos,edge_labels=labels)
plt.show()
```
